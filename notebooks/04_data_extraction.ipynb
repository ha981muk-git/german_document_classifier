```python
import fitz  # PyMuPDF
import os
import re
from PIL import Image
import pytesseract
import io

# -----------------------------
# 3️⃣  FUNCTIONS: EXTRACT + CLEAN
# -----------------------------
"""
def extract_text_from_pdf(pdf_path):
    # Extract text from a PDF file.
    text = ""
    with fitz.open(pdf_path) as doc:
        for page in doc:
            text += page.get_text()
    return text
"""


def extract_text_from_pdf(pdf_path):
    """Extract text from PDF - handles both text and image-based PDFs."""
    text = ""
    with fitz.open(pdf_path) as doc:
        for page_num, page in enumerate(doc):
            # Try text extraction first
            page_text = page.get_text()
            
            if page_text.strip():  # If text found
                text += page_text
            else:  # If no text, try OCR
                print(f"Page {page_num + 1}: No text found, attempting OCR...")
                # Render page as image
                pix = page.get_pixmap(dpi=300)
                img = Image.open(io.BytesIO(pix.tobytes()))
                # OCR the image
                text += pytesseract.image_to_string(img)
    
    return text


def preprocess_text(text):
    """Clean text (same as during training)."""
    text = text.replace("\n", " ")
    text = re.sub(r'<[^>]+>', '', text)
    text = re.sub(r'\s+', ' ', text)
    text = re.sub(r'[^\w\s$€%.,-]', ' ', text).lower().strip()
    return text
```

```python
import fitz  # PyMuPDF

pdf_dir = "../data/data_raw/orders"   # path to your folder of PDFs
pdf_names = []
pdf_texts = []

for file in os.listdir(pdf_dir):
    if file.endswith(".pdf"):
        path = os.path.join(pdf_dir, file)
        pdf_names.append(file)
        raw_text = extract_text_from_pdf(path)
        pdf_texts.append(preprocess_text(raw_text))
```
